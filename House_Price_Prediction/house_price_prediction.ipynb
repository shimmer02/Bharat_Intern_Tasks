```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
```

```python
prices = pd.read_csv('data.csv')
```

```python
prices.head()
```

```python
prices.drop('street', axis='columns', inplace= True)
prices.drop('city', axis='columns', inplace= True)
prices.drop('statezip', axis='columns', inplace= True)
prices.drop('country', axis='columns', inplace= True)
prices.drop('date', axis='columns', inplace= True)
```

```python
prices
```

```python
prices.corr()
```

```python
X = prices.drop('price', axis = 'columns')
```

```python
Y = prices['price']
```

```python
X.describe()
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, random_state=2, test_size= 0.2)
```

```python
X_train.shape
```

```python
reg = LinearRegression()
```

```python
reg.fit(X_train, Y_train)
```

```python
Y_pred = reg.predict(X_train)
mse = mean_squared_error(Y_pred, Y_train)
```
